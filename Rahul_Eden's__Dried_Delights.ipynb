**Eden Dried Delights** is a specialty store that offers a wide selection of dried fruits, conveniently packaged in 100g packs. You have been provided with a dataset containing the sales data of Eden Dried Delights in the below link. This is a cumulative sales of each of the product being sold by the company in a week.

***Task 1: Import the pandas library.***

```python
import pandas as pd
```

***Task 2: Create a DataFrame named 'data' from the given URL and then print the first 5 rows using head() function.***

```python
data = pd.read_csv("https://raw.githubusercontent.com/Invact-Abhay/DOE/refs/heads/main/Eden's%20Dried_Delights%20-%20Data.csv")
data.head()
```

***Task 3: Add a column named 'GrossSales' and calculate the gross sales for each product by multiplying the unit selling price by the units sold.***

```python
data['GrossSales'] = data['UnitSellingPrice'] * data['UnitsSold']
data.head()
```

***Task 4: Add a column named 'DiscountAmount' and calculate the discount amount for each product by multiplying the gross sales by the discount percentage.***

```python
data['DiscountAmount'] = data['GrossSales'] * data['DiscountPercentage']
data.head()
```

***Task 5: Add a column named 'Net Sales' and calculate the net sales for each product by subtracting the discount amount from the gross sales.***

```python
data['Net Sales'] = data['GrossSales'] - data['DiscountAmount']
data.head()
```
